# Methods of Pandas 2

# Pandas fillna()

# fillna fills the NaN values with given values(input)

# Syntax : DataFrame.fillna()

# Parameters of fillna(value, method, axis, inplacee, limit)

```python
import pandas as pd
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1.fillna(0)
```

```python
pd1.fillna(2)
```

```python
pd1.fillna({'Name' : 'none', 'Profit': 0, 'Growth': 0})
```

```python
pd1.fillna(method = 'ffill')

# values is None
# this methos fills the previous values in place of NaN values 
```

```python
pd1.fillna(method = 'pad')

#pad is same as ffill
```

```python
pd1.fillna(method = 'bfill')

#bfill method fills the backward value in place of NaN value
```

```python
pd1.fillna(method = 'ffill' , axis = 0)

# syntax : (method,values)
# without methos this parameter gives error
```

```python
pd1.fillna(method = 'ffill' , axis = 1)

# this fills the previous column values in place of Nan values
```

```python
pd1.fillna(method = 'bfill' , axis = 0)
```

```python
pd1.fillna(method = 'bfill' , axis = 1)

#It fills the backward columns values in NaN values
```

```python
pd1.fillna(0 , limit = 1)

# limit > 0
# limit is use to fill  only specific NaN values in every columns 
# In this parameter we can  fill only one NaN value from every column with 0
# limit(value , method )
```

```python
pd1.fillna(0, limit=2)

# In this parameter we can fill 2 NaN values from every columns with 0
```

```python
pd1.fillna(method = 'ffill', limit =1)

# this parameter sayss that we can fill the previous values in every columns but only One time 
```

```python
pd1.fillna(method = 'ffill', limit =2)

## this parameter sayss that we can fill the previous values in every columns but only two  time 
```

```python
pd1.fillna(5, inplace = True)
pd1
```
