# dropna()

# Parametrs of dropna() method(axis,how,thresh,subset,inplace)

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1.dropna()     

#this method drops all rows and columns which consist NaN values or it remove missing values from dataFrame 
```

```python
pd1.dropna(axis = 1)

# axis = 0   means it remove those row which consist missing values 
#and it is by deafalt case also

#axis = 1  means it remove those columns which consist missing values
```

```python
pd1.dropna(how = 'any')   # or pd1.dropna(axis = 0, how = 'any') because 0 is by default

#any    means if any row and columns have missing values then it drops that row and columns
#any is by default also

#all    means if in out dataFrame all values are Nan values then it drops all values  
```

```python
pd1.dropna(axis = 1,how = 'any')
```

```python
pd1.dropna(how ='all')  #it is for all values

#pd1.dropna(axis = 0, how = 'all')  it drop that row only which consist all NaN values
```

```python
pd1.dropna(axis =1, how = 'all')

# it drop that columns only which consist NaN values
```

```python
pd1.dropna(thresh = 1)

# by this perameter if rows does not have atleast 1 not null values then id drop that rows 
```

```python
pd1.dropna(thresh = 2)

# by this perameter if rows does not have atleast 2 not null values then id drop that rows 
```

```python
pd1.dropna(axis  = 1, thresh =1)

# by this perameter if columns does not have atleast 1 not null values then id drop that columns 
```

```python
pd1.dropna(subset = ['ID']) #those rows which has Nan values in ID coluumn are droped
```

```python
pd1.dropna(subset = ['Name'])   # those rows which has NaN value in Name column are droped
```

```python
pd1.dropna(subset = ['Revenue'])
```

```python
pd1.dropna(subset = ['Profit'])
```

```python
pd1.dropna(inplace = True)   #all the rows which has NaN values are droped 
pd1
```

```python
pd1.dropna(inplace = False)   #all the rows which has NaN values are droped 
pd1
```
