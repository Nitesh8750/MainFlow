# Pandas  Concat() Function

## Pandas provides various facilities (merge, join, concatenate, etc) for easily combining together Series, DataFrame and Panel objects.

## Syntax: Pd.concat()

## parameters: objs, axis, join, join_axis, ignore_index, keys, sort

```python
import pandas as pd
```

```python
sr1 = pd.Series([0,1,2])
sr1
```

```python
sr2 = pd.Series([3,4,5])
sr2
```

```python
pd.concat([sr1, sr2])
```

```python
sr3 = pd.Series([3,4,5,6,7])
sr3
```

```python
pd.concat([sr1,sr3])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4], 'Name': ['A','B','C','D'], 'Class': [5,6,7,8]})
df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8], 'Name': ['E','F','G','H'], 'Class': [9,10,11,12]})
df2
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df2, df1])  
#here axis =0, by default means  DataFrames are concatenate by rows 
```

```python
pd.concat([df2, df1], axis = 1)

#here we concatenate column wise
```

```python
pd.concat([df1, df2], axis = 1)
```

```python
pd.concat([df1, df2], axis = 0)
```

```python
pd.concat([df1, df2], axis = 0, ignore_index = True)
# to correct the index order
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4], 'Name': ['A','B','C','D'], 'Class': [5,6,7,8]})
df1
```

```python
df3 = pd.DataFrame({'ID':[3,4], 'Name': ['C','D'], 'Class': [7,8]})
df3
```

```python
pd.concat([df1, df3])
```

```python
pd.concat([df1, df3], axis = 1)
```

```python
pd.concat([df1, df3], axis = 1, join = 'inner')

# inner means intersection 
# outer means union 
# here by default outer is used
```

```python
pd.concat([df1, df3], axis = 1, join_axes = [df1.index])
```

```python
pd.concat([df1, df3], axis = 1, join_axes = [df2.index])

#both are used like join 
# they are used to return specific dataFrame 
```

```python
pd.concat([df1, df2], keys = [' First df', 'Second df'])

#it is used to label the dataframe 
# it is use to separate the dataframe rows wise 
```

```python
pd.concat([df1, df2], axis =1, keys = [' First df', 'Second df'])

# it is used to separate the dataframe column wise 
```
