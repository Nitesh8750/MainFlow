# Handling Missing Values in Python (na_values,keep_default_na,na_filter)

```python
import pandas as pd
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv", na_values = 'no values')
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv", na_values = 'not available')
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv",  na_values = ['no values', 'not available'])
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv", na_values = {'Industry':'no values' , 'Expenses':'not available'})
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv")
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv" , keep_default_na = False )
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_11.csv" , keep_default_na = True)
pd1
```
