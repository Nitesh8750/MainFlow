# Pandas interpolate()

## Pandas interpolate() function is basically used to fill NaN values in dataframe or series

## syntax : DataFrame.interpolate(), Series.interpolate()
          

## Parameters (method, axis, limit, inplace, limit_direction, limit_area)

```python
pd1.interpolate(limit =1, limit_direction = 'backward')

# by default forward
```

```python
pd1.interpolate(limit =2, limit_direction = 'backward')
```

```python
pd1.interpolate(limit =1, limit_direction = 'both')

# forward and backward
```

```python
#limit_area
#limit_area can be inside , outside 
# by default none
# None : no fill restriction
# inside : only fill NaNs surrounded by valid values (interpolate)
# outside : only fill NaNs outside valid values (extrapolate)
```

```python
pd1.interpolate(limit_area = 'inside')
```

```python
pd1.interpolate(limit_area = 'outside')
```

```python
pd1.interpolate(inplace = True)
pd1

#linear mthod is by default to fill NaNs values
```
