# Perimeter of read_csv

```python
pd1 = pd.read_csv("D:\DATA\Fortune_10.csv", dtype={'ID':'float64'})
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_10.csv", dtype= {'Inception':'float64'})
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_10.csv", dtype = {'Profit':'float64'})
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_10.csv", dtype = {'ID':'float64', 'Profit':'float64'})
pd1
```

```python
pd1 = pd.read_csv("D:\DATA\Fortune_10.csv", true_values= ['Yes'], false_values = ['No'])
pd1
```
