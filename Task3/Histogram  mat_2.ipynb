# Histogram  

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import random
```

```python
ml_student_age = np.random.randint(18,45, (100))
py_student_age = np.random.randint(15,40, (100))
```

```python
print(ml_student_age)
print(py_student_age)
```

```python
plt.hist(ml_student_age)
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]  #category of ML students age on x axis
plt.hist(ml_student_age, bins)
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8)
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, histtype = "bar")
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()

#here orientation is vertical
#  histtype 'bar', 'barstacked', 'step', 'stepfilled'
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal')
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
        color = 'm')
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
        color = 'm', label = 'ML Students')
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
        color = 'm', label = 'ML Students')

plt.hist(py_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'vertical',
        color = 'y', label = 'PY Students')

plt.title("ML & PY Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist(ml_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
        color = 'm', label = 'ML Students')

plt.hist(py_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
        color = 'y', label = 'PY Students')

plt.title("ML & PY Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8,histtype = "bar", orientation = 'vertical',
        color = ["m", "y"], label = ["ML Students", "PY Students"])

#plt.hist(py_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
#        color = 'y', label = 'PY Students')

plt.title("ML & PY Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.show()
```

```python
bins = [15,20,25,30,35,40,45]

plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8,histtype = "bar", orientation = 'vertical',
        color = ["m", "y"], label = ["ML Students", "PY Students"])

#plt.hist(py_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
#        color = 'y', label = 'PY Students')

plt.title("ML & PY Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.show()
```

```python
from matplotlib import style

plt.figure(figsize = (16,9))
plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.8,histtype = "bar", orientation = 'vertical',
        color = ["m", "y"], label = ["ML Students", "PY Students"])

style.use("ggplot")
#plt.hist(py_student_age, bins, rwidth = 0.8,histtype = "bar", orientation = 'horizontal',
#        color = 'y', label = 'PY Students')

plt.title("ML & PY Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.legend()
plt.grid(color = "r", linestyle = "-", linewidth = 2)
plt.show()
```

# Histtypes

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, histtype = "barstacked")
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()

#here orientation is vertical
#  histtype 'bar', 'barstacked', 'step', 'stepfilled'
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, histtype = "step")
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()

#here orientation is vertical
#  histtype 'bar', 'barstacked', 'step', 'stepfilled'
```

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_student_age, bins, rwidth = 0.8, histtype = "stepfilled")
plt.title("ML_Student_Age", fontsize=15)
plt.xlabel("Student Age")
plt.ylabel("No. of Students age")
plt.show()

#here orientation is vertical
#  histtype 'bar', 'barstacked', 'step', 'stepfilled'
```
