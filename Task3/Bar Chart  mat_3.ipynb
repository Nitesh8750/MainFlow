# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#Dataset of Indian Artificial Intelligence Producion(IAIP) Class

# Classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]

classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30,10,20,25,10]  # out of 100 students in each class
class2_students = [40,5,20,20,10]
class3_students = [35,5,30,15,15]
```

```python
plt.bar(classes, class1_students)
```

```python
plt.barh(classes, class1_students)
```

```python
plt.bar(classes, class1_students, width = 0.2)
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", edgecolor = "m")
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5)
```

```python
plt.bar(classes, class1_students, width = 0.2, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9)
```

```python
plt.bar(classes, class1_students, width = 0.5, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":")
```

```python
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students")
plt.legend()
```

```python
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students", visible=False)
plt.legend()
```

```python
style.use("ggplot")
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students")
plt.legend()
```

```python
style.use("ggplot")
plt.figure(figsize = (16,9))
plt.bar(classes, class1_students, width = 0.6, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students")
plt.legend()
```

```python

plt.figure(figsize = (16,9))
plt.barh(classes, class1_students, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students")
plt.legend()
```

```python
plt.figure(figsize = (16,9))
plt.barh(classes, class1_students, align = "edge", color = "y", 
        edgecolor = "m", linewidth = 5, alpha = 0.9, linestyle = ":", label = "Class 1 Students")
plt.title(" Bar Chart of IAIP Class", fontsize = 15)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("NO. Of Students", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))

plt.bar(classes, class1_students,width = 0.2, color = "y", 
        label = "Class 1 Students")

plt.bar(classes, class2_students,width = 0.2, color = "b", 
        label = "Class 2 Students")

plt.title(" Bar Chart of IAIP Class", fontsize = 15)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("NO. Of Students", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students,width, color = "y", 
        label = "Class 1 Students")

plt.bar(classes_index + width, class2_students,width, color = "b", 
        label = "Class 2 Students")

plt.xticks(classes_index + width, classes, rotation = 20)
plt.title(" Bar Chart of IAIP Class", fontsize = 15)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("NO. Of Students", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))

classes_index = np.arange(len(classes))
width = 0.2

plt.bar(classes_index, class1_students,width, color = "y", 
        label = "Class 1 Students")

plt.bar(classes_index + width, class2_students,width, color = "b", 
        label = "Class 2 Students")

plt.bar(classes_index + width + width, class3_students,width, color = "g", 
        label = "Class 3 Students")

plt.xticks(classes_index + width, classes, rotation = 20)
plt.title(" Bar Chart of IAIP Class", fontsize = 15)
plt.xlabel("Classes", fontsize = 15)
plt.ylabel("NO. Of Students", fontsize = 15)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (16,9))

classes_index = np.arange(len(classes))
width = 0.2

plt.barh(classes_index, class1_students,width, color = "y", 
        label = "Class 1 Students")

plt.barh(classes_index + width, class2_students,width, color = "b", 
        label = "Class 2 Students")

plt.barh(classes_index + width + width, class3_students,width, color = "g", 
        label = "Class 3 Students")

plt.yticks(classes_index + width, classes, rotation = 20)
plt.title(" Bar Chart of IAIP Class", fontsize = 15)
plt.ylabel("Classes", fontsize = 15)
plt.xlabel("NO. Of Students", fontsize = 15)
plt.legend()
plt.show()
```
