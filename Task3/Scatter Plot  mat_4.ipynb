```python
df_google_play_store_apps = pd.read_csv("D:\DATA\googleplaystore.csv", nrows = 1000)
df_google_play_store_apps.shape  #shape means (row, columns)
```

```python
df_google_play_store_apps
```

```python
x = df_google_play_store_apps["Rating"]
y = df_google_play_store_apps["Reviews"]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r")
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*")
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100)
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5)
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5, linewidth = 10)
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5, linewidth = 10,
           edgecolor = "g")
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviews", fontsize = 13)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha = 0.5, linewidth = 10,
           edgecolor = "g")
plt.scatter(x,df_google_play_store_apps["Installs"], c = "y", marker = "o", s = 100, alpha = 0.5, linewidth = 10,
           edgecolor = "c")
plt.title("Google Play Store Apps", fontsize = 15)
plt.xlabel("Rating", fontsize = 13)
plt.ylabel("Reviewss & Installs", fontsize = 13)
```
