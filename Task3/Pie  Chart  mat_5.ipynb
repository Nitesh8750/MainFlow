```python
classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
class1_students = [45, 15, 25, 35, 30]  # out of 100 students in class

explode = [0,0,0.2,0,0]
colors = ["c",'y','b','r','g']
textprops = {"fontsize":15}
wedgeprops = {"linewidth": 5, "width":1, "edgecolor":"k"}
plt.figure(figsize = (16,9))
plt.pie(class1_students, labels = classes, explode = explode, colors = colors,
       autopct = "%0.2f%%", shadow = True, radius = 1.4, startangle = 270,
       textprops = textprops, counterclock = True, wedgeprops = wedgeprops,
       center= (0,1), frame= True, rotatelabels = True,pctdistance = 0.7, 
        labeldistance = 1.1)
plt.legend()
plt.show()
```

# Some more types of pie charts 

```python
#classes = ["Python", "R", "Artificial Intelligence", "Machine Learning", "Data Science"]
#class1_students = [45, 15, 25, 35, 30]  # out of 100 students in class

explode = [0,0,0.2,0,0]
colors = ["c",'y','b','r','g']
textprops = {"fontsize":15}

plt.figure(figsize = (16,9))
wedgeprops = {"linewidth": 4, "width":2, "edgecolor":"k"}

plt.pie(
    class1_students, 
    labels = classes,
    labeldistance = 1.4,
    explode = explode, 
    colors = colors,
    autopct = "%0.2f%%",
    pctdistance = 0.6,
    shadow = True,
    radius = 2.0,
    startangle = 270,
    textprops = textprops,
    counterclock = True,
    wedgeprops = wedgeprops,
    center= (0,1),
    frame= True,
    rotatelabels = True)
plt.legend()
plt.show()
```

```python
wedgeprops = {"linewidth": 3, "width":3, "edgecolor":"k"}
plt.figure(figsize = (16,9))
plt.pie(
    class1_students, 
    labels = classes,
    labeldistance = 1.0,
    explode = explode, 
    colors = colors,
    autopct = "%0.2f%%",
    pctdistance = 0.6,
    shadow = True,
    radius = 2.0,
    startangle = 270,
    textprops = textprops,
    counterclock = True,
    wedgeprops = wedgeprops,
    center= (0,0),
    rotatelabels = True,
    frame= True)
plt.legend()
plt.show()
```
